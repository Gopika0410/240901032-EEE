```python
import pandas as pd

# Create DataFrame
df = pd.DataFrame({
    'Animal': ['Tiger', 'Lion', 'Cheetah', 'Deer'],
    'Max speed': ['130', '110', '180', '160']
})

# Convert 'Max speed' to numeric
df['Max speed'] = df['Max speed'].astype(int)

# Group by 'Animal' and sum
b = df.groupby(['Animal']).sum()

print("\nOriginal DataFrame\n", df)
print("\nGroup by through Column using sum\n", b)
```

```python
import pandas as pd

# Create DataFrame
df = pd.DataFrame({
    'Animal': ['Tiger', 'Lion', 'Cheetah', 'Deer'],
    'Max speed': [130, 110, 180, 160]  # Make sure numeric
})

# Group by 'Animal' and count
c = df.groupby(['Animal']).count()

print("\nOriginal DataFrame\n", df)
print("\nGroup by through Column using count\n", c)
```

```python
import pandas as pd

# Create DataFrame
df = pd.DataFrame({
    'Animal': ['Tiger', 'Lion', 'Cheetah', 'Deer'],
    'Max speed': [130, 110, 180, 160]  # Numeric
})

# Group by 'Animal' and take the first row of each group
d = df.groupby(['Animal']).first()

print("\nOriginal DataFrame\n", df)
print("\nGroup by through Column using first\n", d)
```
