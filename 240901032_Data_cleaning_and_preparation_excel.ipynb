```python
import pandas as pd
pd.set_option('display.max_columns',10)
print("Original data:\n")
df=pd.read_excel("/content/ppml excel 07 bt.xlsx")
print(df)
```

```python
f=df['Name'].str.strip()
f.to_excel("/content/dcap.xlsx")
print("\nReplace value:")
k=df.fillna(method='pad')
print(k)
print("\nUpdated list\n")
k.to_excel("/content/dcap.xlsx")
```

```python
print("\nDrop one particular column and its values:\n")
if 'social' in k.columns:
    k.drop(['social'],axis=1,inplace=True)
else:
    print("Column 'social' already dropped or does not exist.")
print(k)
print("\nUpdated list..\n")
k.to_excel("/content/dcap.xlsx")
```

```python
print("\nDrop NaN rows:\n")
df=pd.read_excel("/content/dcap.xlsx")
print(df)
x=df.dropna()
print(x)
print("\nUpdated data list:\n")
x.to_excel("/content/dcap.xlsx")
```

```python
print("\nReplace values:\n")
n=pd.read_excel("/content/dcap.xlsx")
print("\nOriginal data:\n")
print(n)
y=n.replace({89:88})
print("\nReplaced dataset:\n")
print(y)
print("\nUpdated dataset:\n")
y.to_excel("/content/dcap.xlsx")
```
