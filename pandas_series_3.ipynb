```python
import pandas as pd
import numpy as np
#creating seies from array
arr=np.array(['P','a','n','d','a','s'])
a=pd.Series(arr)
print("Series from array:")
print(a)
```

```python
#Series from dictionary
arr={'x':0.,'y':1.,'z':2.}
b=pd.Series(arr)
print("\n\nSeries from dictionary:\n")
print(b)
```

```python
#series using scalar
x=pd.Series(4,index=[0,1,2,3])
print("\nSeries using scalar\n")
print(x)
```

```python
#Series through index
x=pd.Series([1,2,3],index=['a','b','c'])
print("\nSeries through index:\n")
print(x)
```

```python
a=pd.Series(data=[1,2,3,4])
print("\n a series:\n",a)
print("\n Index:\n",a.index)
print("\n Value:\n",a.values)
print("\n Shape:",a.shape)
print("\n Dimension:",a.ndim)
print("\n Size:",a.size)
```
