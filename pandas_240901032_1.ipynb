```python
import pandas as pd
#empty data frame
df=pd.DataFrame()
print(df)
```

```python
emp=pd.Series(['Parker','John','Smith','William'])
id=pd.Series([1102,107,109,114])
frame={'Emp':emp,'ID':id}
result=pd.DataFrame(frame)
print("\nSeries to  DataFrame")
print(result)
```

```python
print("\nExtraction one colum:\n")
print(result['Emp'])
```

```python
print("\nAdding new column:\n")
result['Age']=pd.Series([35,24,40,38])
print(result)
```

```python
print("\nExtracting the third row:\n")
print(result.loc[2])
```

```python
print("\nDeleting one column:\n")
del result['Age']
print(result)
```

```python
#Slice rows
print("\nSlice rows:\n",result[1:3])
```

```python
d2= pd.DataFrame([['Dale',123],['Mark',143]],columns=['Emp','ID'])
result=pd.concat([result,d2], ignore_index=True)
print("\n Adding new row values:\n", result)
```

```python
print("\nDeleting particular row:\n",result.drop(1))
```
