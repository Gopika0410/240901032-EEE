```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt

# Create dataset
data = {
    'Transmitter': ['T1','T2','T1','T3','T2','T1','T3','T2','T1','T3'],
    'Strength': [75,50,90,40,85,70,30,60,80,20],
    'Frequency': [2000,1500,2200,1300,2100,1800,1200,1600,1900,1100],
    'Quality': ['Good','Bad','Good','Bad','Good','Good','Bad','Bad','Good','Bad']
}

df = pd.DataFrame(data)

print(df)

# Encoding categorical values
d1 = {'T1':0, 'T2':1, 'T3':2}
df['Transmitter'] = df['Transmitter'].map(d1)

d2 = {'Good':1, 'Bad':0}
df['Quality'] = df['Quality'].map(d2)

# Features and target
features = ['Transmitter', 'Strength', 'Frequency']
X = df[features]
y = df['Quality']

# Create model
model = DecisionTreeClassifier()

# Train model
model.fit(X, y)

# Plot tree
tree.plot_tree(model, feature_names=features)
plt.show()
```
